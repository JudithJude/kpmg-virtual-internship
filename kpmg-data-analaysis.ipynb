```python
data = pd.ExcelFile("/kaggle/input/halima-data/KPMG_VI_New_raw_data_update_final (2).xlsx")
```

```python
#loading each excel sheet
cus_demo = pd.read_excel(data, "CustomerDemographic", header = 1)
cus_addr = pd.read_excel(data, "CustomerAddress", header = 1)
trans = pd.read_excel(data, "Transactions", header = 1)
new_cus_list = pd.read_excel(data, "NewCustomerList", header = 1)
```

```python
#dropping null values in the customer demographic sheet
cus_demo.dropna()
```

```python
#checking each column in customer demographic
cus_demo["gender"].head(50)
```

```python
#the gender column isn't consistent, so some corrections will be made
cus_demo["gender"].replace(["F", "Femal"], "Female", inplace = True)
cus_demo["gender"].replace(["M"], "Male", inplace = True)
cus_demo["gender"].replace(["U"], "Undecided", inplace = True)

cus_demo["gender"].head()
#since there are no nulls, we can move ahead
```

```python
#the dob column
cus_demo["DOB"].head()
```

```python
#creating a column to define each customer's age since we can't really tell by looking
#at their date of birth
import datetime as dt
cus_demo["age"] = (dt.datetime.now() - cus_demo["DOB"]) / np.timedelta64(1, "Y")

cus_demo.head()
```

```python
#I noticed there are nan and inf values in the age column, so I had to replace and drop
#them because i can't covert to integer(Cannot convert non-finite values (NA or inf) to integer)
cus_demo = cus_demo.replace([np.inf, -np.inf], np.nan).dropna(axis = 0)

cus_demo.head()
```

```python
#creating an age category so it can be easily visualised later
cus_demo["age_cat"] = ((round(cus_demo["age"] / 10)) * 10).astype(int)

cus_demo.head()
```

```python
#finally, we drop the default column since it just has random characters and no sense
#can be made out of it.
#cus_demo.drop(["default"], axis = 1)
cus_demo.drop('default',axis = 1, inplace = True)
```

```python
#the state column is not consistent and it has to be corrected
cus_addr["state"].replace(["New South Wales"], "NSW", inplace = True)
cus_addr["state"].replace(["Victoria"], "VIC", inplace = True)

cus_addr.head()
```

```python
#dropping the null values
trans.dropna(inplace = True)
trans.isnull().sum()
```

```python
#checking there are duplicates
trans.duplicated().sum()
```

```python
#the info showed that product_first_sold_date as a float instead of a date. Correcting...
trans["product_first_sold_date"] = pd.TimedeltaIndex(trans["product_first_sold_date"],
                                                    unit = "d") + dt.datetime(1900, 1, 1)

trans.head()
```

```python
#dropping the null values
new_cus_list.dropna(inplace = True)
new_cus_list.isnull().sum()
```

```python
###Merging the four cleaned sheets
```

```python
old_cus_list = pd.merge(cus_demo, cus_addr, on = "customer_id")
old_cus_list.head()
```

```python
#merging the new list with the transaction sheet to enable me get the recency, frequency,
#and monetary values
old_cus_list = pd.merge(old_cus_list, trans, on = "customer_id")
old_cus_list.head()
```

```python
#checking the latest transaction date
trans["transaction_date"].max() 
```

```python
#Since the latest transaction occurred at 2017-12-31, we'll be using it to calculate the
#Recency values
day = "2017-12-30"
day = pd.to_datetime(day)
recency = trans.groupby("customer_id").agg({"transaction_date" : lambda x:
                                            ((day-x.max()).days)})
#creating the recency column
recency = recency.loc[:,["transaction_date"]]
recency =  recency.rename(columns={'transaction_date': 'recency'})
recency.head()
```

```python
#adding the recency column to the ex_cus_list_tr sheet
old_cus_list = old_cus_list.merge(recency, on = "customer_id", how = "left")
old_cus_list.head()
```

```python
old_cus_list.isnull().sum()
```

```python
old_cus_list["recency"].fillna(old_cus_list["recency"].mean())
old_cus_list.head()
```

```python
#getting the frequency values
frequency = trans.groupby("customer_id").count()
frequency = frequency.loc[:,["transaction_id"]]
frequency = frequency.rename(columns = {"transaction_id" : "frequency"})
frequency.head()
```

```python
#merging the frequency column to the ex_cus_list_tr sheet
old_cus_list = old_cus_list.merge(frequency, on = "customer_id", how = "left")
old_cus_list.head()
```

```python
#For the monetary values
monetary = trans.loc[trans["order_status"] == "Approved"].groupby(
    "customer_id").agg({"list_price" : "sum", "standard_cost" : "sum"})
monetary["monetary"] = monetary["list_price"] - monetary["standard_cost"]
monetary = monetary.drop(["list_price", "standard_cost"], axis = 1)
monetary.head()
```

```python
#merging the monetary column to the old_cus_list
old_cus_list = old_cus_list.merge(monetary, on = "customer_id", how = "left")
old_cus_list.head()
```

```python
#we'll be working with both the new and old customer lists
#adding age to both sheets
old_cus_list["age"] = (dt.datetime.now() - old_cus_list["DOB"]) / np.timedelta64(1, "Y")
old_cus_list.head()
```

```python
#adding age category
old_cus_list["age_category"] = ((round(old_cus_list["age"] / 10)) * 10).astype(int)

old_cus_list.head()
```

```python
#adding age and the category for new customer list
new_cus_list["age"] = (dt.datetime.now() - new_cus_list["DOB"]) / np.timedelta64(1, "Y")
new_cus_list.head()
```

```python
new_cus_list["age_category"] = ((round(new_cus_list["age"] / 10)) * 10).astype(int)

new_cus_list.head()
```
